```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
```

#### Description from Krackhardt Kite graph method

```python
The traditional labeling is: Andre=1, Beverley=2, Carol=3, Diane=4, Ed=5, Fernando=6, Garth=7, Heather=8, Ike=9, Jane=10.
```

#### Create Input File for Relationship Mappings

```python
#guidance for input creation: https://towardsdatascience.com/customizing-networkx-graphs-f80b4e69bedf
rels = pd.read_csv('https://raw.githubusercontent.com/jforster19/Data620/main/kite_graph_input%20-%20Sheet1.csv')
```

```python
rels.head()
```

### Create Kite Graphic to mirror assignment prompt

```python
G = nx.from_pandas_edgelist(rels, 'From', 'To', create_using=nx.Graph())
positions = {'Ed':(5,5),'Beverly':(0,10),'Andre':(0,15),'Garth':(10,10),'Diane':(5,12.5),'Carol':(5,17.5),\
             'Fernando':(10,15),'Heather':(20,12.5),'Ike':(25,12.5),'Jane':(30,12.5)}
nx.draw_networkx(G, pos=positions,with_labels=True, edgecolors='blue', node_color='gray', node_size=1000)
```
